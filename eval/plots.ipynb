# Script to generate plots & evaluate the methods

```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
with open('eval_data_exercise.pickle', 'rb') as handle:
    eval_data = pickle.load(handle)
```

## The average runtime and memory usage for methods joblib and pickle using the exercise data

```python
runtime_joblib = [t[1] for t in eval_data["runtime"]["joblib"]]
runtime_pickle = [t[1] for t in eval_data["runtime"]["pickle"]]

memory_joblib = [t[1] for t in eval_data["memory_usage"]["joblib"]]
memory_pickle = [t[1] for t in eval_data["memory_usage"]["pickle"]]

# Print the averages with comments
print(f"Average runtime for joblib: {np.average(runtime_joblib)}")
print(f"Average runtime for pickle: {np.average(runtime_pickle)}")
print()
print(f"Average memory usage for joblib: {np.average(memory_joblib)}")
print(f"Average memory usage for pickle: {np.average(memory_pickle)}")
```

```python
# Create the plots
plt.figure(figsize=(12, 6))

# Subplot 1: Memory usage comparison
plt.subplot(2, 1, 1)
plt.scatter(range(len(memory_pickle)), memory_pickle, label="Pickle Memory Usage (KB)", color="blue", alpha=0.7, s=10)
plt.scatter(range(len(memory_joblib)), memory_joblib, label="Joblib Memory Usage (KB)", color="green", alpha=0.7, s=10)
plt.title("Scatter Plot of Memory Usage: Pickle vs Joblib")
plt.ylabel("Memory (KB)")
plt.xlabel("Iteration")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Create the plots
plt.figure(figsize=(12, 6))

# Subplot 2: Runtime comparison
plt.subplot(2, 1, 1)
plt.scatter(range(len(runtime_pickle)), runtime_pickle, label="Pickle Runtime", color="blue", alpha=0.7, s=10)
plt.scatter(range(len(runtime_joblib)), runtime_joblib, label="Joblib Runtime", color="green", alpha=0.7, s=10)
plt.title("Scatter Plot of the Runtime: Pickle vs Joblib")
plt.ylabel("ms")
plt.xlabel("Iteration")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
sorted_runtime_joblib = sorted(eval_data["runtime"]["joblib"], key=lambda x: x[1])
sorted_runtime_pickle = sorted(eval_data["runtime"]["pickle"], key=lambda x: x[1])
sorted_memory_joblib = sorted(eval_data["memory_usage"]["joblib"], key=lambda x: x[1])
sorted_memory_pickle = sorted(eval_data["memory_usage"]["pickle"], key=lambda x: x[1])

def print_top_5(label, sorted_list):
    print(f"{label}:")
    for item in sorted_list[-5:]:
        print(f"  {item[0]} - Time: {item[1]:.4f} ms" if "runtime" in label else f"  {item[0]} - Memory: {item[1]:.4f} KB")
    print()


print_top_5("Top 5 runtimes for joblib", sorted_runtime_joblib)
print_top_5("Top 5 runtimes for pickle", sorted_runtime_pickle)
print_top_5("Top 5 memory usages for joblib", sorted_memory_joblib)
print_top_5("Top 5 memory usages for pickle", sorted_memory_pickle)
```
