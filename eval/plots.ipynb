# skript to generate plots & evaluate training etc

```python
import matplotlib.pyplot as plt
```

```python
# Load the memory and timing data from the file
pickle_memory_usage = []
joblib_memory_usage = []

with open("memory_time_usage.txt", "r") as f:
    lines = f.readlines()
    
    # Extract memory usage for Pickle
    pickle_index = lines.index("Pickle Memory Usage (KB):\n")
    pickle_memory_usage = list(map(float, lines[pickle_index + 1].strip().split(",")))
    
    # Extract memory usage for Joblib
    joblib_index = lines.index("Joblib Memory Usage (KB):\n")
    joblib_memory_usage = list(map(float, lines[joblib_index + 1].strip().split(",")))
```

```python
# Create the plots
plt.figure(figsize=(12, 6))

# Subplot 1: Memory usage comparison
plt.subplot(2, 1, 1)
plt.scatter(range(len(pickle_memory_usage)), pickle_memory_usage, label="Pickle Memory Usage (KB)", color="blue", alpha=0.7, s=10)
plt.scatter(range(len(joblib_memory_usage)), joblib_memory_usage, label="Joblib Memory Usage (KB)", color="green", alpha=0.7, s=10)
plt.title("Scatter Plot of Memory Usage: Pickle vs Joblib")
plt.ylabel("Memory (KB)")
plt.xlabel("Iteration")
plt.legend()

plt.tight_layout()
plt.show()
```
